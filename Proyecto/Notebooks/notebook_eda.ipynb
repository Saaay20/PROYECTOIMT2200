```python
import pandas as pd


try:
    df_a_visualizar = pd.read_csv("..\Data\Raw\Dataset_viviendas_final.csv")
    print("Dimensiones del DataFrame original:", df_a_visualizar.shape)

    # Calcular los límites para las columnas numéricas
    limite_precio = df_a_visualizar['precio_uf'].quantile(0.98)
    limite_m2 = df_a_visualizar['m2_totales'].quantile(0.93)
    limite_m2_construidos = df_a_visualizar['m2_construidos'].quantile(0.98)
    limite_antiguedad = df_a_visualizar['antiguedad_anos'].quantile(0.98)
    limite_estacionamientos = df_a_visualizar['estacionamientos'].quantile(0.98)
    limite_banos = df_a_visualizar['banos'].quantile(0.98)
    
    df_a_visualizar["comuna"] = (
        df_a_visualizar["comuna"]
        .astype(str)                            
        .str.strip()                            
        .str.replace(r"\s+", " ", regex=True)   
        .str.replace(u"\xa0", " ")              
        .str.title()                            
    )

    # Aplicar todos los filtros juntos
    df_filtrado = df_a_visualizar[
        (df_a_visualizar['precio_uf'] > 0) &
        (df_a_visualizar['precio_uf'] < limite_precio) &
        (df_a_visualizar['m2_totales'] > 0) &
        (df_a_visualizar['m2_totales'] < limite_m2) &
        (df_a_visualizar['m2_construidos'] > 0) &
        (df_a_visualizar['m2_construidos'] < limite_m2_construidos)&
        (df_a_visualizar['antiguedad_anos'] < limite_antiguedad) &
        (df_a_visualizar['estacionamientos'] < limite_estacionamientos) &
        (df_a_visualizar['banos'] < limite_banos) &
        
        
        
        (df_a_visualizar["comuna"].notna()) &
        (df_a_visualizar["comercios"].notna()) &
        (df_a_visualizar["condominio_cerrado"].notna()) &
        (df_a_visualizar["jardin"].notna()) &
        (df_a_visualizar["direccion"].notna()) &
        (df_a_visualizar["titulo"].notna()) &
        (df_a_visualizar["url"].notna()) &
        (df_a_visualizar["salud"].notna()) &
        (df_a_visualizar["piscina"].notna()) &
        (df_a_visualizar["tipo_vivienda"].notna()) &
        (df_a_visualizar["educacion"].notna()) &
        (df_a_visualizar["quincho"].notna())
        

    ]

    print("Dimensiones del DataFrame filtrado:", df_filtrado.shape)

    
    df_filtrado.to_csv("..\Data\Procesados\Dataframe_final.csv", index=False)
    print("\nDataFrame filtrado guardado exitosamente como 'Dataframe_final.csv'.")

except FileNotFoundError:
    print("Error: No se encontró el archivo 'Dataset_viviendas_final.csv'. Por favor, asegúrate de que esté en la misma carpeta.")

```

## Limpieza, estandarización y filtrado del dataset

### 1️⃣ Carga del dataset
Se cargó el archivo `Dataset_viviendas_final.csv` desde la carpeta de datos procesados.  
Antes de realizar cualquier análisis exploratorio, fue necesario **limpiar, estandarizar y filtrar** las variables para reducir el impacto de valores atípicos y errores de captura.

---

### 2️⃣ Aplicación de percentiles para filtrado de valores extremos

Para evitar que los **outliers** distorsionaran las visualizaciones y análisis posteriores, se utilizaron **percentiles** como límites superiores en las variables numéricas.  
De esta forma, se eliminan los valores más extremos, manteniendo la gran mayoría de los datos representativos del mercado inmobiliario.

| Variable | Percentil aplicado | Justificación |
|-----------|-------------------|----------------|
| `precio_uf` | 0.98 | Excluye el 2 % de propiedades más caras, que suelen ser atípicas o de lujo. |
| `m2_totales` | 0.93 | Recorta el 7 % superior para mantener propiedades de tamaño estándar. |
| `m2_construidos` | 0.98 | Filtra el 2 % superior de construcciones muy grandes o poco frecuentes. |
| `antiguedad_anos` | 0.98 | Elimina valores de antigüedad anómalos o registros mal ingresados. |
| `estacionamientos` | 0.98 | Filtra casos con cantidades excesivas de estacionamientos (poco comunes). |
| `banos` | 0.98 | Evita distorsiones causadas por viviendas con demasiados baños, típicamente no representativas. |



---

### 3️⃣ Limpieza y estandarización de texto (columna `comuna`)
La columna `comuna` presentaba inconsistencias (mayúsculas/minúsculas, espacios, caracteres especiales).  
Se normalizó aplicando:

- Conversión a texto (`astype(str)`)
- Eliminación de espacios duplicados (`str.strip()` y `str.replace(r"\s+", " ", regex=True)`)
- Remoción de caracteres no separables (`str.replace(u"\xa0", " ")`)
- Capitalización de formato (`str.title()` → *La Reina, Lo Prado, Las Condes…*)

Esto asegura una **representación uniforme** para análisis y visualizaciones posteriores.

---

### 4️⃣ Eliminación de filas con valores nulos en columnas clave
Se eliminaron registros con valores nulos en campos esenciales para el análisis, como:
- `comuna`, `direccion`, `titulo`, `url`
- `tipo_vivienda`, `educacion`, `salud`, `comercios`
- `jardin`, `piscina`, `quincho`, `condominio_cerrado`

Estas columnas son críticas para la interpretación del dataset, y mantener filas incompletas afectaría la calidad de los gráficos y correlaciones.

---

### 5️⃣ Resultado del filtrado

Tras aplicar todos los filtros:
- El dataset pasó de **6232 a 5339 filas** .
- Se eliminó ruido y valores atípicos sin perder representatividad.
- Se generó un nuevo archivo limpio:  
  **`Dataframe_final.csv`**

---

### 🧠 Conclusión general

El proceso de limpieza y filtrado mediante **percentiles**, junto con la **normalización de texto** y la **eliminación de valores nulos**, permite obtener un dataset más confiable, coherente y visualmente interpretable.  
Esto garantiza que los análisis exploratorios y visualizaciones posteriores reflejen **tendencias reales del mercado inmobiliario**, evitando sesgos causados por datos extremos o inconsistentes.

# Análisis Exploratorio de Datos (EDA) del Mercado Inmobiliario – Región Metropolitana

## 1️⃣ Propósito general
El objetivo del EDA fue **explorar y comprender el comportamiento de los precios de viviendas** en la Región Metropolitana, utilizando como fuente el dataset `Dataset_viviendas_final.csv` tras un proceso de limpieza y filtrado basado en percentiles.  
El análisis busca identificar **patrones de variación en el precio (en UF)** según factores estructurales (m², baños, dormitorios), temporales (antigüedad), geográficos (comuna) y de equipamiento (amenities como piscina, quincho, condominio, jardín, etc.).

---

## 2️⃣ Preparación y limpieza del dataset
Se aplicaron filtros utilizando cuantiles para remover **outliers extremos** que distorsionaban la escala de los gráficos.  
Los límites usados fueron:
- `precio_uf` ≤ percentil 98  
- `m2_totales` ≤ percentil 93  
- `m2_construidos`, `antiguedad_anos`, `baños`, `estacionamientos` ≤ percentil 98  

Además:
- Se normalizaron nombres de comunas (`str.title()`, eliminación de espacios, caracteres no imprimibles).  
- Se eliminaron registros con valores nulos en variables clave (`comuna`, `tipo_vivienda`, `educacion`, `salud`, `comercios`, `amenities`, etc.).  

Esto garantizó un dataset limpio, homogéneo y representativo del mercado inmobiliario real.

---

## 3️⃣ Visualizaciones y hallazgos principales

### 🔹 3.1 Relación entre precio y superficie construida
Los primeros gráficos de dispersión (`precio_uf` vs `m2_construidos` y `precio_uf` vs `m2_totales`) muestran una **relación positiva**:  
las viviendas más grandes tienden a tener un mayor precio absoluto.  
No obstante, la dispersión vertical indica que existen **otros factores** (ubicación, tipo, equipamiento) que también influyen en el valor.

El filtrado por percentiles mejoró significativamente la visualización, evitando que los outliers arrastraran la escala.

---

### 🔹 3.2 Relación entre precio y antigüedad
El gráfico `precio_uf` vs `antiguedad_anos` muestra que, en general, las viviendas **más nuevas son más costosas**.  
Sin embargo, se observa una concentración importante entre 0 y 20 años de antigüedad, correspondiente a proyectos recientes del mercado inmobiliario actual.  
No se detectó una relación lineal perfecta.
---

### 🔹 3.3 Dispersión general del precio
El gráfico de dispersión del precio en función del índice de observación confirma una **alta variabilidad**, con clústeres de precios intermedios y algunos peaks que representan viviendas de lujo.  
La mayoría de las propiedades se concentra bajo las 10.000 UF, lo que concuerda con el rango medio del mercado.

---

### 🔹 3.4 Matriz de correlación
La matriz de correlación evidencia:
- Correlaciones **fuertes y positivas** entre `precio_uf`, `m2_construidos`, `m2_totales` y `baños`.  
- Correlaciones **moderadas** con `dormitorios` y `estacionamientos`.  
- Correlación **negativa** entre `antiguedad_anos` y el resto, reflejando el efecto de depreciación del tiempo.
